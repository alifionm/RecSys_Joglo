```python
import pandas as pd
import re # Membuat regex (membersihkan tanda baca)

# Import Sastrawi untuk bahasa Indonesia
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat objek stopword remover dan stemmer
factory_stop = StopWordRemoverFactory()
stopword_remover = factory_stop.create_stop_word_remover()

factory_stem = StemmerFactory()
stemmer = factory_stem.create_stemmer()

print("Library NLP Sastrawi siap digunakan!")
```

```python
df_wisata = pd.read_csv('Dataset Clean/Data_Wisata_Clean.csv')

print("=== SEBELUM DIBERSIHKAN ===")
print(df_wisata['Deskripsi_Wisata'].head(3))
```

```python
def clean_text(text):
    
    text = str(text)
    
    # Case Folding
    text = text.lower()
    
    # Cleansing
    text = re.sub(r'[^a-z\s]', ' ', text)
    
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Stopword Removal
    text = stopword_remover.remove(text)
    
    # Stemming
    text = stemmer.stem(text)
    
    return text

test_kalimat = "Keindahan Candi Prambanan di Jogja sangat MEMUKAU! Pengunjung 100% puas."
print("Hasil Test:", clean_text(test_kalimat))
```

```python
print("Sedang memproses teks... Tunggu sebentar ya...")
df_wisata['Deskripsi_Clean'] = df_wisata['Deskripsi_Wisata'].apply(clean_text)

print("\n=== SESUDAH DIBERSIHKAN ===")
print(df_wisata[['Deskripsi_Wisata', 'Deskripsi_Clean']].head(3))

df_wisata.to_csv('Dataset Clean/Data_Wisata_Preprocessed.csv', index=False)
print("\nDataset berhasil disimpan: Data_Wisata_Preprocessed.csv")
```

```python
import mlflow

mlflow.set_experiment("Preprocessing_Joglo")

with mlflow.start_run(run_name="NLP_Cleaning_Sastrawi"):
    # Log parameter
    mlflow.log_param("library", "Sastrawi")
    mlflow.log_param("stemming", True)
    
    # Log metrik
    mlflow.log_metric("total_rows", len(df_wisata))
    
    # Log file hasil
    mlflow.log_artifact("Dataset Clean/Data_Wisata_Preprocessed.csv")
    
    print("Eksperimen berhasil dicatat ke MLflow!")
```
