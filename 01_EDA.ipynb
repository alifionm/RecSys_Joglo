```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Load data
df_wisata = pd.read_csv('Dataset Clean/Data_Wisata_Clean.csv')
df_rating = pd.read_csv('Dataset Clean/Data_Rating_Clean.csv')

# Cek shape
print("Shape data wisata:", df_wisata.shape)
print("Shape data rating:", df_rating.shape)
```

```python
# Hitung Sparsity Level

n_users = df_rating['User_ID'].nunique()
n_items = df_wisata['ID_Wisata'].nunique()
n_ratings = len(df_rating)

total_possible = n_users * n_items
sparsity = (1.0 - (n_ratings / total_possible)) * 100

print(f"Total Unik User: {n_users}")
print(f"Total Unik Tempat Wisata: {n_items}")
print(f"Total Interaksi (Rating): {n_ratings}")
print(f"Tingkat Sparsity: {sparsity:.2f}%")
```

```python
# Cek sebaran rating user
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df_rating, x='Rating', hue='Rating', palette='viridis', legend=False)

plt.title('Distribusi Rating Wisatawan (Skala 1-5)', fontsize=14, pad=15)
plt.xlabel('Rating (Bintang)', fontsize=10)
plt.ylabel('Jumlah Ulasan', fontsize=10)

# Tambah label angka di atas bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Mapping kategori wisata
kategori_count = df_wisata['Kategori'].value_counts()

plt.figure(figsize=(10, 6))

kategori_count.plot(kind='bar', color='coral', edgecolor='black')

plt.title('Jumlah Destinasi Berdasarkan Kategori', fontsize=14, pad=15)
plt.xlabel('Kategori Wisata', fontsize=12)
plt.ylabel('Jumlah Tempat', fontsize=12)

plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show()
```
